```python
%matplotlib inline
%config InlineBackend.figure_format='retina'

import glob
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import vaex

from matplotlib.lines import Line2D

sns.set(style='ticks', 
        rc={'font.family'          : 'serif',
            'text.usetex'          : True,
            'axes.axisbelow'       : True,
            'mathtext.fontset'     : 'cm',
            'mathtext.default'     : 'regular',
            'xtick.direction'      : 'in',
            'ytick.direction'      : 'in',
            'xtick.major.width'    : 1.1,
            'ytick.major.width'    : 1.1,
            'axes.labelpad'        : 5.0,
            'legend.framealpha'    : 0.95,
            'legend.handlelength'  : 1.2,
            'legend.handletextpad' : 0.5,
       })

pal = sns.color_palette('colorblind')
dark = sns.color_palette('dark')
sns.set_palette(pal)
```

```python
df = pd.read_csv('triout_m33_frg_260sqrarcmin.dat', delim_whitespace=True,
                 ).rename(columns=lambda x: x.replace('mag','').replace('W1','W'))
```

```python
fig = plt.figure(figsize=(9,6))
gs = fig.add_gridspec(4, 6)
ax1 = fig.add_subplot(gs[0:2, 0:2])
ax2 = fig.add_subplot(gs[0:2, 2:4])
ax3 = fig.add_subplot(gs[0:2, 4:6])
ax4 = fig.add_subplot(gs[2:4, 1:3])
ax5 = fig.add_subplot(gs[2:4, 3:5])

axes = [ax1, ax2, ax3, ax4, ax5]

plot_kwargs = dict(s=1, color='k', alpha=0.5)

ax1.scatter(df.F275W-df.F336W, df.F336W, **plot_kwargs)
ax1.set_xlim(-1.5, 2.5)
ax1.set_ylim(26, 16)
ax1.set_xlabel('F275W--F336W')
ax1.set_ylabel('F336W')
ax2.scatter(df.F336W-df.F475W, df.F475W, **plot_kwargs)
ax2.set_xlim(-2.5, 4)
ax2.set_ylim(28, 18)
ax2.set_xlabel('F336W--F475W')
ax2.set_ylabel('F475W')
ax3.scatter(df.F475W-df.F814W, df.F814W, **plot_kwargs)
ax3.set_xlim(-1.5, 6.5)
ax3.set_ylim(28.5, 17.5)
ax3.set_xlabel('F475W--F814W')
ax3.set_ylabel('F814W')
ax4.scatter(df.F475W-df.F160W, df.F160W, **plot_kwargs)
ax4.set_xlim(-1.5, 11)
ax4.set_ylim(26.5, 16)
ax4.set_xlabel('F475W--F160W')
ax4.set_ylabel('F160W')
ax5.scatter(df.F110W-df.F160W, df.F160W, **plot_kwargs)
ax5.set_xlim(-1.5, 2.5)
ax5.set_ylim(26.5, 16)
ax5.set_xlabel('F110W--F160W')
ax5.set_ylabel('F160W')

fig.tight_layout()
fig.savefig('survey_paper_figures/trilegal_foreground.pdf', dpi=300, bbox_inches='tight')
```
